Dans cette partie de code, nous voulons créer un dataframe qui correspond aux informations des joueurs regroupées par années. 

```python
import pandas as pd
```

```python
path='.\Data\Data_utiles\info_matches.csv'
info_matches1=pd.read_csv(path)
columns_needed = ['l_1stIn', 'l_svpt', 'l_1stWon', 'l_2ndWon', 'w_svpt', 'w_1stWon', 'w_2ndWon','tourney_date', 'loser_rank', 'winner_rank','loser_rank','w_bpFaced','l_bpFaced','l_bpSaved','w_bpSaved','l_SvGms','w_SvGms']

    # Filtrer le DataFrame pour exclure les lignes avec des NaN dans les colonnes spécifiées
info_matches=info_matches1.dropna(subset=columns_needed)
```

```python
joueurs_dict = {}


for index, row in info_matches.iterrows():
    winner_id = row['winner_id']
    loser_id = row['loser_id']

    if winner_id not in joueurs_dict:
        joueurs_dict[winner_id] = {'name': row['winner_name'], 'hand': row['winner_hand'], 'height': row['winner_ht']}
        for year in range(1993, 2023):  
            joueurs_dict[winner_id][str(year)] = []

    if loser_id not in joueurs_dict:
        joueurs_dict[loser_id] = {'name': row['loser_name'], 'hand': row['loser_hand'], 'height': row['loser_ht']}
        for year in range(1993, 2023):  
            joueurs_dict[loser_id][str(year)] = []

print(joueurs_dict)
```

```python
for joueur in joueurs_dict.keys():
    année=1993
    matchs=0
    win=0
    pourc_serv_win_pnt = 0
    pourc_return_win_pnt = 0
    pourc_serv_in=0
    pourc_break_games=0
    pourc_serv_games_win=0
    pourc_break_point_saved=0
    pourc_break_point_made=0
    mean_ranking_oppo=0

    for i in range(len(info_matches)):
        tourney_date_str = info_matches['tourney_date'].iloc[i]
        if str(tourney_date_str)[:4] == année:
        
            if info_matches.winner_id[i]==joueur and  info_matches.l_svpt[i]!=0 and int(info_matches.w_svpt[i])!=0 and int(info_matches.w_SvGms[i])!=0 and int(info_matches.w_bpFaced[i])!=0:
                pourc_serv_in+= int(info_matches.w_1stIn[i]) / int(info_matches.w_svpt[i])
                pourc_serv_win_pnt += (int(info_matches.w_1stWon[i]) + int(info_matches.w_2ndWon[i])) / int(info_matches.w_svpt[i])
                pourc_return_win_pnt+= (int(info_matches.l_svpt[i]) - int(info_matches.l_1stWon[i]) - int(info_matches.l_2ndWon[i])) / int(info_matches.l_svpt[i])
                pourc_serv_games_win+= (int(info_matches.w_SvGms[i]) - ( int(info_matches.w_bpFaced[i])-int(info_matches.w_bpSaved[i])))/int(info_matches.w_SvGms[i])
                pourc_break_games+= 1 - (int(info_matches.l_SvGms[i]) - ( int(info_matches.l_bpFaced[i])-int(info_matches.l_bpSaved[i])))/int(info_matches.l_SvGms[i])
                mean_ranking_oppo+=info_matches.loser_rank[i]
                matchs+=1
                win+=1
                pourc_break_point_saved+=int(info_matches.w_bpSaved[i])/int(info_matches.w_bpFaced[i])
                pourc_break_point_made+=1-int(info_matches.l_bpSaved[i])/int(info_matches.l_bpFaced[i])
            elif info_matches.loser_id[i] == joueur and info_matches['w_svpt'][i] != 0 and int(info_matches['w_svpt'][i]) != 0 and int(info_matches['w_SvGms'][i]) != 0 and int(info_matches['w_bpFaced'][i]) != 0:
                pourc_serv_in += int(info_matches['l_1stIn'][i]) / int(info_matches['l_svpt'][i])
                pourc_serv_win_pnt += (int(info_matches['l_1stWon'][i]) + int(info_matches['l_2ndWon'][i])) / int(info_matches['l_svpt'][i])
                pourc_return_win_pnt += (int(info_matches['w_svpt'][i]) - int(info_matches['w_1stWon'][i]) - int(info_matches['w_2ndWon'][i])) / int(info_matches['w_svpt'][i])
                pourc_serv_games_win += (int(info_matches['l_SvGms'][i]) - (int(info_matches['l_bpFaced'][i]) - int(info_matches['l_bpSaved'][i]))) / int(info_matches['l_SvGms'][i])
                pourc_break_games += 1 - (int(info_matches['w_SvGms'][i]) - (int(info_matches['w_bpFaced'][i]) - int(info_matches['w_bpSaved'][i]))) / int(info_matches['w_SvGms'][i])
                mean_ranking_oppo += info_matches['winner_rank'][i]
                matchs += 1
                pourc_break_point_saved += int(info_matches['l_bpSaved'][i]) / int(info_matches['l_bpFaced'][i])
                pourc_break_point_made += 1 - int(info_matches['w_bpSaved'][i]) / int(info_matches['w_bpFaced'][i])
            else:
                pass
        else:
            if matchs>0:
                win=win/matchs
                pourc_serv_win_pnt = pourc_serv_in/matchs
                pourc_return_win_pnt = pourc_serv_in/matchs
                pourc_serv_in=pourc_serv_in/matchs
                pourc_break_games=pourc_break_games/matchs
                pourc_serv_games_win=pourc_serv_games_win/matchs
                pourc_break_point_saved=pourc_break_point_saved/matchs
                pourc_break_point_made=pourc_break_point_made/matchs
                mean_ranking_oppo=mean_ranking_oppo/matchs
                joueurs_dict[joueur][str(année)]=[matchs,win,pourc_return_win_pnt,pourc_return_win_pnt,pourc_break_games,pourc_break_point_made,pourc_break_point_saved,pourc_serv_games_win,pourc_serv_in]
                année+=1
                matchs=0
                win=0
                pourc_serv_win_pnt = 0
                pourc_return_win_pnt = 0
                pourc_serv_in=0
                pourc_break_games=0
                pourc_serv_games_win=0
                pourc_break_point_saved=0
                pourc_break_point_made=0
                mean_ranking_oppo=0
                

            else:
                joueurs_dict[joueur][str(année)]=[matchs,win,pourc_return_win_pnt,pourc_return_win_pnt,pourc_break_games,pourc_break_point_made,pourc_break_point_saved,pourc_serv_games_win,pourc_serv_in]
                année+=1
                

           

print(info_matches[101990])   

```
