```python
import pandas as pd
import matplotlib.pyplot as plt



 
```

Dans cette partie de code, nous voulons récupérer le pourcentage de 1er service in, ainsi que les poins gagné derrière le service ainsi que les points gagnés au retour 

```python
df=pd.read_csv('./data/atp_matches_1994.csv')
```

```python
# Colonnes nécessaires pour les calculs
columns_needed = ['l_1stIn', 'l_svpt', 'l_1stWon', 'l_2ndWon', 'w_svpt', 'w_1stWon', 'w_2ndWon', 'loser_rank', 'winner_rank']

# Filtrer le DataFrame pour exclure les lignes avec des NaN dans les colonnes spécifiées
df_filtered = df.dropna(subset=columns_needed)

print(len(df_filtered))
```

on va noter pour_cserv_win10 et pour_cserv_win50 le pourcentage de point gagné au service pour un joueur entre le top 40 et 50.

on va noter pourc_return_win10 et pourc_return_win50.
on va noter pourc_serv_in_10 et pourc_serv_in_50.


```python

df=pd.read_csv('./Data/atp_matches_2022.csv')

# Colonnes nécessaires pour les calculs
columns_needed = ['l_1stIn', 'l_svpt', 'l_1stWon', 'l_2ndWon', 'w_svpt', 'w_1stWon', 'w_2ndWon', 'loser_rank', 'winner_rank']

# Filtrer le DataFrame pour exclure les lignes avec des NaN dans les colonnes spécifiées
df_filtered = df.dropna(subset=columns_needed)

print(len(df_filtered))


a = 0  # Initialize counters
b = 0

pour_cserv_win10_sum = 0
pour_cserv_win50_sum = 0
pourc_return_win10_sum = 0
pourc_return_win50_sum = 0
pourc_serv_in_10_sum=0
pourc_serv_in_50_sum=0
for i in range(len(df_filtered)):
    try:
        if df_filtered.loser_rank[i] in range(0, 11) and df_filtered.l_svpt[i]!=0 and int(df_filtered.w_svpt[i])!=0:
            a += 1
            pourc_serv_in_10 = df_filtered.l_1stIn[i] / df_filtered.l_svpt[i]
            pour_cserv_win10 = (df_filtered.l_1stWon[i] + df_filtered.l_2ndWon[i]) / df_filtered.l_svpt[i]
            pourc_return_win10 = ( df_filtered.w_svpt[i] - df_filtered.w_1stWon[i] - df_filtered.w_2ndWon[i]) / (df_filtered.w_svpt[i])
            pour_cserv_win10_sum += pour_cserv_win10
            pourc_return_win10_sum += pourc_return_win10
            pourc_serv_in_10_sum+=pourc_serv_in_10
        elif df_filtered.loser_rank[i] in range(40, 51) and df_filtered.l_svpt[i]!=0 and int(df_filtered.w_svpt[i])!=0:
            b += 1
            pourc_serv_in_50 = int(df_filtered.l_1stIn[i]) / int(df_filtered.l_svpt[i])
            pour_cserv_win50 = (int(df_filtered.l_1stWon[i]) + int(df_filtered.l_2ndWon[i])) / int(df_filtered.l_svpt[i])
            pourc_return_win50 = ( + int(df_filtered.w_svpt[i]) - int(df_filtered.w_1stWon[i]) - int(
                df_filtered.w_2ndWon[i])) /  int(df_filtered.w_svpt[i])
            pour_cserv_win50_sum += pour_cserv_win50
            pourc_return_win50_sum += pourc_return_win50
            pourc_serv_in_50_sum+=pourc_serv_in_50
        elif df_filtered.winner_rank[i] in range(0, 10) and df_filtered.l_svpt[i]!=0 and int(df_filtered.w_svpt[i])!=0:
            a += 1
            pourc_serv_in_10 = int(df_filtered.w_1stIn[i]) / int(df_filtered.w_svpt[i])
            pour_cserv_win10 = (int(df_filtered.w_1stWon[i]) + int(df_filtered.w_2ndWon[i])) / int(df_filtered.w_svpt[i])
            pourc_return_win10 = (int(df_filtered.l_svpt[i]) - int(df_filtered.l_1stWon[i]) - int(
                df_filtered.l_2ndWon[i])) / int(df_filtered.l_svpt[i]) 
            pour_cserv_win10_sum += pour_cserv_win10
            pourc_return_win10_sum += pourc_return_win10
            pourc_serv_in_10_sum+=pourc_serv_in_10
        elif df_filtered.winner_rank[i] in range(40, 51) and df_filtered.l_svpt[i]!=0 and int(df_filtered.w_svpt[i])!=0:
            b += 1
            pourc_serv_in_50 = int(df_filtered.w_1stIn[i]) / int(df_filtered.w_svpt[i])
            pour_cserv_win50 = (int(df_filtered.w_1stWon[i]) + int(df_filtered.w_2ndWon[i])) / int(df_filtered.w_svpt[i])
            pourc_return_win50 = (int(df_filtered.l_svpt[i]) - int(df_filtered.l_1stWon[i]) - int(
                df_filtered.l_2ndWon[i])) / int(df_filtered.l_svpt[i]) 
            pour_cserv_win50_sum += pour_cserv_win50
            pourc_return_win50_sum += pourc_return_win50
            pourc_serv_in_50_sum+=pourc_serv_in_50
    except KeyError:
        # Handle the KeyError (key not present) gracefully, for example, by skipping the iteration
        pass

if a > 0:
    pour_cserv_win10_avg = pour_cserv_win10_sum / a
    pourc_return_win10_avg = pourc_return_win10_sum / a
    pourc_serv_in_10_avg=pourc_serv_in_10_sum/a
else:
    pour_cserv_win10_avg = 0
    pourc_return_win10_avg = 0

if b > 0:
    pour_cserv_win50_avg = pour_cserv_win50_sum / b
    pourc_return_win50_avg = pourc_return_win50_sum / b
    pourc_serv_in_50_avg = pourc_serv_in_50_sum /b
else:
    pour_cserv_win50_avg = 0
    pourc_return_win50_avg = 0

stat=(pour_cserv_win10_avg, pour_cserv_win50_avg, pourc_return_win10_avg, pourc_return_win50_avg,pourc_serv_in_10_avg,pourc_serv_in_50_avg)
stat_list.append(stat)
```

```python

l_pour_cserv_win10_avg = [element[0] for element in stat_list]
l_pour_cserv_win50_avg = [element[1] for element in stat_list]
l_pourc_return_win10_avg=[element[2] for element in stat_list]
l_pourc_return_win50_avg=[element[3] for element in stat_list]
l_pourc_serv_in_10_avg=[element[4] for element in stat_list]
l_pourc_serv_in_50_avg=[element[5] for element in stat_list]
# Années de 1993 à 2022
years = list(range(1993, 2023))

# Création du graphe
plt.figure(figsize=(10, 6))  # Taille du graphe

plt.plot(years, l_pour_cserv_win10_avg, label='Variable 1')  # Tracer variable1
plt.plot(years,l_pour_cserv_win50_avg , label='Variable 2')  # Tracer variable2

plt.xlabel('Années')  # Étiquette de l'axe des x
plt.ylabel('Valeurs')  # Étiquette de l'axe des y
plt.title('Évolution des variables au fil des années')  # Titre du graphe
plt.legend()  # Ajout de la légende
plt.grid(True)  # Afficher la grille

# Afficher le graphe
plt.show()

A=stat_list
```
